```python

train_numerical= train.select_dtypes(exclude='object')
train_categorical = train.select_dtypes(include='object')

```

```python
train_categorical.info()
```

```python
train_numerical.info()
```

```python
## standerizem  numerical_data amb StandardScaler
## primer substituim tots els valors nulls per la mitjana
for k in train_numerical:
    if train_numerical[k].isnull().values.any() == True:
        train_numerical[k] = train_numerical[k].fillna(round(train_numerical[k].mean()))
        
```

```python
sc = StandardScaler()
continuous_vars =train_numerical[['DepTime','CRSDepTime','ArrTime','CRSArrTime','FlightNum','ActualElapsedTime','CRSElapsedTime', 'AirTime', 'ArrDelay', 'DepDelay', 'Distance','TaxiIn', 'TaxiOut', 'CarrierDelay','WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']]
train[continuous_vars.columns]= sc.fit_transform(continuous_vars)
```

```python
fe = train.groupby('DayofMonth').size()/len(train)
train.loc[:,'Day_freq_encode'] = train['DayofMonth'].map(fe)
```

```python
# convertim a One Hot Encoding  la columna UniqueCarrier
dum = pd.get_dummies(train['UniqueCarrier'])
train = pd.concat([train, dum], axis=1)
```

```python
train.drop('UniqueCarrier',inplace=True,axis=1)
```
